```python
# importing libraries
import speech_recognition as sr
import os
from pydub import AudioSegment
from pydub.silence import split_on_silence

# create a speech recognition object
r = sr.Recognizer()

# a function to recognize speech in the audio file
# so that we don't repeat ourselves in other functions
def transcribe_audio(path):
    # use the audio file as the audio source
    with sr.AudioFile(path) as source:
        audio_listened = r.record(source)
        # try converting it to text
        try:
            text = r.recognize_google(audio_listened)
        except sr.UnknownValueError as e:
            text = ""  # Set an empty string if recognition fails
            print("Error:", str(e))
    return text

# a function that splits the audio file into chunks on silence
# and applies speech recognition
def get_large_audio_transcription_on_silence(path):
    """Splitting the large audio file into chunks
    and apply speech recognition on each of these chunks"""
    # open the audio file using pydub
    sound = AudioSegment.from_file(path)
    # split audio sound where silence is 500 miliseconds or more and get chunks
    chunks = split_on_silence(sound,
                              # experiment with this value for your target audio file
                              min_silence_len=500,
                              # adjust this per requirement
                              silence_thresh=sound.dBFS - 14,
                              # keep the silence for 1 second, adjustable as well
                              keep_silence=500,
                              )
    folder_name = "audio-chunks"
    # create a directory to store the audio chunks
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)
    whole_text = ""
    # process each chunk
    for i, audio_chunk in enumerate(chunks, start=1):
        # export audio chunk and save it in
        # the `folder_name` directory.
        chunk_filename = os.path.join(folder_name, f"chunk{i}.wav")
        audio_chunk.export(chunk_filename, format="wav")
        # recognize the chunk
        text = transcribe_audio(chunk_filename)
        text = f"{text.capitalize()}. "
        print(chunk_filename, ":", text)
        whole_text += text
    # return the text for all chunks detected
    return whole_text

# Example usage:
audio_file_path = "/content/output.wav"
result_text = get_large_audio_transcription_on_silence(audio_file_path)
print("Transcription result:")
print(result_text)
```

```python
import spacy
from collections import Counter
from heapq import nlargest
from spacy.lang.en.stop_words import STOP_WORDS

# Load the English language model in spaCy
nlp = spacy.load('en_core_web_sm')

# Define the function to extract important points from text
def extract_key_points(text, num_sentences=5):
    # Parse the text using spaCy
    doc = nlp(text)

    # Extract named entities from the text
    entities = [entity.text for entity in doc.ents]

    # Extract keywords from the text
    keywords = [token.text for token in doc if not token.is_stop and not token.is_punct and token.pos_ != 'PRON']

    # Combine the named entities and keywords into a single list
    key_points = entities + keywords

    # Count the frequency of each key point
    freq = Counter(key_points)

    # Calculate the weight of each key point based on its frequency and position in the text
    weights = {key: freq[key] / len(doc) for key in freq}

    # Calculate the score of each sentence based on the weights of its key points
    sentence_scores = {sentence: sum(weights.get(word, 0) for word in sentence.split()) for sentence in text.split('.')}

    # Extract the top 'num_sentences' sentences with the highest scores
    summary_sentences = nlargest(num_sentences, sentence_scores, key=sentence_scores.get)

    # Format the summary sentences as a numbered list
    summary = ''
    for i, sentence in enumerate(summary_sentences):
        summary += f'{i+1}.{sentence.strip()}.\n'

    return summary

# Example text
text="It is now 6:30 and i'll be calling the meeting of the ridgefield city council to order will you please rise and join us in the flag salute. Cancel chipman. . Play 937. . . . . Miss julie will you please call raul. . . Mr stuart are there any late changes to the agenda. . Are we moving up an appointment. Yes mayor we are thank you for the reminder about that could we please move roman numeral 92 under mayor the appointment confirmation of the parks board member up to. . Roman numeral 2. Item four. Perfect thank you very much. Okay so with that we will be making an appointment. We will be making an appointment to the parks board position number 7. Do i have a motion. Do i have a second. . All those in favor. . Motion carries mr earl are you here with us this evening i think he was going to try and attend did he make it. Okay so. He is quite involved i'm a lot of non-profit work. And he's very passionate about work in the outdoors and. He will be the incoming director of the clark county parks foundation as well so he's going to bring a wealth of experience so we are excited to have him. Okay so moving on to. The next i'm a business this evening we have a very special item the life-saving award chief dorian. . . Mayor lindsay council members and citizens. A police officer job is quite often it focuses on enforcement accountability and justice. But our uniform and badge symbolize more than authority. They also reflect empathy. Care and the potential to make a life-saving difference. Tonight we are recognizing two of our ridgefield officers whose dedication to duty and swift actions in a medical crisis saved a life. If officer watson an officer connecticut come to the front please. . . On june 12th 2023 in the early morning hours. Officer craig watson an officer rob cuneta were dispatched to an aed call. The call notes indicated to patient a 73 year old female was unconscious and not breathing. Officer watson and officer cuneta arrived on scene within 2 minutes. When they entered the residents they met the reporting party and they found the patient lying on the floor inside the home. Officer watson and officer conetta immediately started cpr. And they attached an ad to the patient. When the aed indicated no shock. Officers resume cpr. Clark county fire and rescue arrived on scene and took over the cpr responsibilities. As a result of the officers actions. And the actions of personnel from clark county fire and rescue. The patient sold signs of responsiveness trying to be prior to being transported to the hospital by ambulance. It is my honor to say that the patient survive this medical incident. I would like to introduce you to chris drump christine drum. And her significant other kenneth sanders. Officer watson in oxford connecticut quick response to the scene and their immediate efforts with cpr contributed to to christine survival. As a result of their experience training and quick response to this emergency. They were able to take action to save this life. Officer watson and officer kennedy's efforts exemplify the high standards of the law enforcement profession. And their actions are a credit to all who serve as agents of public safety. On behalf of the citizens of ridgefield. And the ridgefield police department. It is my pleasure and honor to recognize your actions with this lifesaving award for your outstanding professional service. Julia if you could bring up the. . . Robert. I don't know if there's anybody else that would like to address or if you would like to make any statements you're welcome to. . . I just want to say it's an honor working in the city of richfield serving the citizens. . It's a great city. Can do it without the help of my brothers and sisters in blue thank you to my family. For always supporting me. We love you guys. . Super happy. . What what a wonderful department to equip us in training so well and what a great city to. To be so supportive and and a special thanks to the fire guys who also got there really quick so it's not all not all craig and i the fire guys did a wonderful job as well. Thank you. So we're going to take like a 2-minute break because. I know all of you are here for our meeting right not for the life-saving award so i'm going to take a 2-minute break so we can get some photos and if you want to say you are welcome to stay. . . . . . . . . . . . . . . . . . . Okay it looks like. We're ready to move on with our meeting on the next item of business this evening is public comment. Anyone wishing to address the city council regarding items not subject to a specific public hearing may come forward at this time to the podium or use the raised hand feature located on the toolbar at the bombing resume screen. For controls over the phone press star 94 raise hand and press star 6 to mute un. Please state your name and limit your comments to three minutes is there anyone wishing to give public comment please come on up to the podium. Green arrow and dean maldonado hair. I would like to suggest that the council and the city start taking a look at. Meetings it's in some of the other towns we have an annual. Blue light. . And i've come across two or three things that have been. A little confusing a little irritating and one of the things we want to do is we want to work as developers with the city. To make it easier to grow and get through the system. And one of the things is you know chuck doesn't need me you know calling him up every other day about this or that or. In ordinances are just like the rotary phone some of them have had their time and they need to be updated. So one of the things they do is they have the annual. Developer city meeting. To lead to that point we used to call it the hateful eight the eight things i hated. . And we would send that into the city. And the city engineer in the planning would look at it and go okay. To make things go smoother and faster. I suggest that members of the council maybe the city. Put together a group and reach out to the developers. Ordinances or however we could do it because this actually became a very successful thing and they hold it. Natalie. . But there are obstacles out there one of the unfortunate things is we have weather here. . . . . How to get into a pre-application and it gets held up then the weather is here we can't build the next spring well now you're asking us for architectural plans on top of the site plans which shouldn't even be there in my opinion. And now i got to spend another $188. 180.00 and i'm not there to the spring. So it's not your fault it's not my fault it's the process that we need to look at and make sure it's updated just like the rotary phone. Stop signs in the red flashing lights with roundabouts so that's what i would suggest that you look. Thank you. Thank you is there anyone else wishing to get public comment this evening please come on up to the podium. Play state your name my name is alex earl i was just. . And i look forward to serving the community and. Give me my advice for it might be helpful thank you so much. . Is there anyone else wishing to get public comment this evening. Mr lee is there anyone online. Not seeing anyone online not having anyone else. In person or online we will be moving forward to the consent agenda. Madame are i moved to approve the consent agenda as presented. . . We have emotion to approve the consent agenda as presented in a second is there any discussion council. . Opposed. . All right this evening we have a presentation from the salary commission we have an update from the chair sean ford. . Good evening just a quick refresher to begin about the salary commission in april 2014 the city of council. Establish an independent salary commission the role of commission is to review. Establish the salaries of the mayor in the council members. Salaries that are established by the commission are final. The commission is made up of five citizens within the ridgefield city limits or is urban growth boundary. Commissioners are appointed by the mayor and subject to approval of the council. No commission member may be a city officer official or employee of the city or any of their immediate family members. The commission we met on june 13th of this year and reviewed and discussed comparable salaries of mayors and council members of comparable size cities. We did not receive any public comment at that meeting or prior to the meeting. After deliberation the commission voted to implement an increase effective january 1st 2024 as follows. For the mayor on monthly salary of $1,350 for an annual maximum of $16,200. Salary of $675 for an annual maximum of 8. And that was the business of the commission. Thank you very much i appreciate your service thank you for coming this evening. Okay the next item on the agenda is business item number one the second reading of ordinance number 1407 the lottie property or muahh removal is left. Thank you mary good evening council members of the public this is the second reading of the proposed zone change removing the ridgefield mixed use overlay. From the lodi property on pioneer street i'll flip here to our vicinity map. Thank you julie. All right so the subject property is. On the south side of pioneer street between royal road and 56. Sean highlighted on the maps the map on the right shows the current zoning designation which is employment-based zoning with the ridgefield mixed use overlay on northern roughly 2/3. Or about 22 acres of the property. The proposal is to maintain the employment based on but remove the ridgefield mixed use overlay and its entirety. And the purpose of this is to facilitate future development. Along pioneer street including a medical clinic in the northeast corner of the site. Cancel held a public hearing and discussion on this topic. 2 weeks ago at your last meeting there were clarifying questions but no major issues to look into so at this point i will go ahead and hand it back to you for any further discussion and questions and emotion thank you. Do you have any additional questions. At this time. . Not hearing any questions i would ask miss julie to please read the ordinance into the record thank you mary ordinance number 1407 ordinance of the city of richland washington amending the zoning of assessor's parcel. 213989000 by removing the ritual mixed use zoning overlay. I would accept a motion. Anime i moved to. Except. . 14 ordnance number 1407 is. Pres. . We have a motion to approve. Presented will you please call rollins. . . The next item of business this evening is a motion to the award of construction contract for the royal road project south 15th street to south 5th way mr green. Thank you mayor council and honored guests i'm here to. Present for motion for a word of the construction contract for the royal road. Project south 15th street to south 5th way. . Been a little bit of a journey getting here i'm not going to cover every single part of the journey. But i'm going to cover a little bit of the highlight so what part of this this award. Sell railroad is the next phase. North of the. . About. . . Will north of 15. And wells drive. So with this will include is widening and improvements to railroad. . Both north of the. Endpoint at 15th. And we're back filling the landscaping. In sidewalk park south of 15th that you. Restored to this project. Complete the quarter all the way up to. . I'm not going to get into lighting engineering details but we have taken a lot of lessons learned from the the project that we are wrapping up now. To the south. And some of those you'll see here one is first phase is what's called the joint utility trench. And that means and you're seeing the first part of that now with clark public utilities kind of getting stuff moved. To the sides we can actually get in and start. Developing a trench and putting the overground overhead utilities in on the underground. . . That's going to be the first phase and that will happen this fall. . . The lesson learned here is we're not going to start road work until we actually get the utility trench completed so we can actually do a kind of clean. Just wait for the contractor to start working. First phase will be construction on the west side. But we keep the east side open to traffic. . Second phase will it be to build the east side or we keep well we use the new west side. And if you notice. The prevailing theme here is we're keeping this open as much as we can. Without any closures without a foreclosures needed of the routes that's a staging. . Stage 4 is kind of wrapping it up and fortunately. . I didn't show what the other part of stage 4 was. Or stage 1 depending on how the contractor wants to work. But that would be the landscaping and sidewalk south the 15th street in wells drive. To fill in that part that was not included in the current project so that will be up to the contract if they want to do early. . But it will be done. So the other thing that you've probably heard about. Is another lesson learned as we are now applying for the first time what's called supplemental bidder responsibility criteria. What that means is we've got criteria that we are applying to the contractors that are the apparent little bitters. To make sure they're qualified. An experience in doing the work similar to what we're trying to do here on royal road. And so there's never questions that we asked. And that is to look at three projects that they successfully completed in the last 5 years or similar. That there hasn't been a lot of change orders that have increased the cost or cause delay on the project. Their experience with staging of of work so we don't have closures. And with a joint utility trench. . Also recognize we wanted to make sure we didn't. Scare away the bidders who are contractors who might be new. Better using people who worked for other contractors and similar projects in a substantive role so. We actually allowed contractors to turn in. Those projects that were similar in nature and we actually got six bids is part of this so we did not scare away better. . Stop with the slides there but as you know. We did have an experience with our. Little bitter. Did not turn in a required form it was acquired under state law. They protested that and through that and. . Illegal hearing that was last week. That contractor was determined to be non-responsive and was disqualified from the bidding process from there. The remaining responsive and responsible little bitter was western united civil group. And their bed was 4, 9 4 5 _. We are recommending that that would be the award of the contractor who would receive the award plus. We would. Request the council add an additional construction contingency of 250.00 that's about 5%. Unforeseen circumstances. On the project to get that completed. We do have money in this year's budget we are planning to request money in next year's budget to wrap this project up. It also this year's budget includes an action that you'll be addressing next council meeting that will be for construction management consultant. To assist the city staff in. Ensuring that this project is has inspection and support to make sure it gets done. In a timely manner. . We are requesting action on the word and i'm going to stop my presentation and return it back to the mayor. Thank you very much mr green are there any questions. . Yes i have a couple questions in the past presentations you had talked about the three phases mr green and that it was roughly going to take 18 months i believe is what you said for the entire project in october and wrapping up with late fall next year. Okay so i'm in the. My question is on the supplemental bitter criteria. Page 110 in our packet. It says the contractor must be able to demonstrate cost and schedule control in their project submitted for criterion one. Not to me that. Who owns is there a critical path schedule that shows you know this can't be done till the next thing is done is does a contractor have that and who takes ownership of that to monitor it as it the phases. Are are completed if one is taking a little longer obviously you can't go to the next one so who controls that. There's a couple things one is the cannot start any of the road work until the giant utility trenches completed. And so what the contractor will be doing is submitting a critical path schedule for each phase as we we go into it. That will be monitored with what's called a three week look ahead. So every week we will meet with the contractor and we have a construction management consultant we just had interviews. I made a recommendation where to negotiations with them now if somebody who. Is very experienced these type of projects including joint utility trenches an amazing amount of experience with that and stag. Construction so they very much. Know how to read. And maintain a critical pass schedule it'll be up to the contractor. To provide the critical pass schedule and maintain that it'll be up to us the city and our construction management consultant to make sure that that is followed. We we go through them. Benefit of critical pass schedule for thursday night or used to what that means is we have certain tasks that have to happen in a sequence. And the schedule is dependent upon. Addressing them in that sequence. I think other things can happen but it's these items that really affect the schedule. And if one of those items starts falling behind. Then we need to work on what's called a change or corrective management plan. Which is how we going to catch up at the end if if this thing starts. Going longer than it should. Okay thank you very much for that explanation. Cancel verbal. Mr green during the project you suggested that you'll do one lane at a time. One sided one second so how will traffic be managed on the open side during the construction. So we'll probably want to reduce the construction speed you probably saw the signs when i came on board in in january first thing i wanted to do on the projecting up. Just make sure we get 25 mile an hour signs out there. How to keep traffic slow and that's probably what we'll do on this. Also will probably have to tighten the road up a bit just because it's not a wide road now and they're going to be working. Right next to it. But typically what we want to do is make sure if there's adequate traffic signs and cones out there to make sure people know. Are there supposed to be going. And there might be times where. If there's a driveway or a side street that would have to cross. A work zone we might have to detour at the side street traffic around that to keep those people safe. But for the most part we want to make sure that. There's adequate. Control devices out there to make sure the people know which lane they're supposed to go in which sides are supposed to be. And want to keep a little bit of a buffer. From that traffic and the workers obviously want to keep our workers safe as well. . . Mr green's comments and i. Where we were talking about cross traffic. And. Clover hill. Entrance to. . Railroad there a person could probably. Clothes that off completely. And everybody could go out the other. . I'm not trying to tell you what happened to be very beneficial. Do the contractor. . Sou. Yeah and that's a great point because we're going to make sure that if we're going to close an entrance people have another way out. Fieldcrest is one of those that will be closed because. If you come over royal if you're heading south you notice there's a hill and then a dip in the site distances is pretty substandard if you're trying to get out on the royal there we're going to cut that down. And in cunning it down we actually have to cut fieldcrest down a little bit with the median. To match things up so points were going to have to do that it'll be clear that there's. The other way out on 215th street during that time. Thank you. . . Construction project comes with the you know impacts to the public but how much downtime or how much total road closure can we expect. For this leg of a royal road during the construction. Well our ideal goal. We're starting out is zero full closures. Now it's not meaning that there won't be any closures at all. There are times we're going to need to transition between stages and in that transition we're going to need to do some paving and probably. Either try to have direct traffic around that but we might need to do something maybe on a weekend to. After a day to make that transition but the intent is we're not going to have. Weeks long or month-long closures like we experience on south royal so we're talking like a day or two at a time potentially thank you. I had a question that kind of follows up on counselor. Buckles. . . Safety along the corridor while it's under construction and when the last section was done. . When you were in the middle of it you kind of jigged and jagged and it was at times unclear where you were supposed to go and it's kind of a dark section of road and when you're going into your thing is going to start in october. Long dark. Sections of. You know when the sun goes down early what are we doing anything to improve. . Lighting along the corridor during construction. . Possibility as we've got the car puppy utilities poles along there now we could string some. Call cobra headlights on onto them temporarily. I'm going to come up so we have some power there and so we get the permanent light fixtures in. Thank you mary part. One other safety concern that you might want to look at mr green is that mr wells mentioned maybe closing also 10th you know you get the main entrance and then you have 10th. If 10th is used as an alternate to get in and out of as you're coming from rosauers and you go to make that right hand turn on 10th. There's blackberry bushes that are really high this a blind spot you cannot see. So what happens is when you make a right on 10th there's somebody coming out and they're not in their proper position and there could be a traffic accident right there so if you're going to use 10. I have friends that live in clover hill and i go there quite often. And so and when i'm traveling with my friends they tell me oh don't turn that way go to the main entrance and i know why it's because of the risk you cannot see if somebody's coming out and if they're in their proper lane. So please take a look at the visibility there if we're going to use 10 as the main in and out. Yeah i will definitely take a look and make sure the contractor has that kept open. What are the benefits of having roundabouts at both ends of the project. Is if we need to restrict the left turns. We can use the roundabouts to make u-turns to then come around to do the right turns in. Which then takes out that left turn conflicts that all. In in the constrained area that we're working with will will probably be a little tough to make those left turns at times. Thank you counselor chapman. Spell mr greenwich i already wrote mr stewart this note but because i was so critical of the way the other contracts were written i have to tell you that. I was thrilled when i read this bid package how you wrote it i mean it's just. So complete and you just crossed every day and daughter every i. What we need to have to know that this is going to be a very well done project so. Kudos to you for. For the way that this was handled thank you thank you. Cancel are there any other questions or comments. At this time. Okay so i'm not hearing any others i would accept emotion. I moved to award the contract for the royal road south 15th street. To south 5th way project to western united civil group llc from yacolt washington and the amount not to exceed 4,945. 1,434 and 17 cents with an additional construction contingency of 250,000. Second. Is there any further discussion council. . . Award the railroad project to western united civil group. For the amount of 400. 4,945 4434.17 for the additional construction contingency of 250.00. All those in favor. . . . All right we got that one done. The next item of business is a motion to select the vendor for the roundabout grapes were at that time of year already mr stewart. In late 2014. The city council decided to work with volunteers in a local winemaker to plant marichelle first grapes in the 56th and 65th roundabouts on pioneer. That effort was to help to better promote the viticultural area around ridgefield and within it. Between that time and 2021. Cancel had multiple conversations about the specific. Uses for those grapes and a purpose for them as well as a process for how those grapes may be used. So the city council memorialize that process in 2021 for determining the use of the roundabout grapes each year the first step in that process is where we are today which is to research possible vendors who would be willing to work with the city on creating a product from the grapes which will remain as public property for non pecuniary basically. Nonprofit uses. In 2023 city staff reached out to winemakers around clark county to determine interest in working with the city to store process and ultimately bottle wine from the marriage hall. Roundabout grapes direct outreach was also undertaken by our city councilors. Leading to contact an interest by 14 acres winery owner dave reg. In the staff report it says that no other winery reply to the request we did get a reply this morning i got a reply from bethany vineyard they very much appreciated the outreach in the interest and they were interested in doing something with the city but not this year. It's something they just find themselves overwhelmed with kind of the work that they have in front of them they said we'd love to talk with you in a future you're about that. So if city council chooses to proceed with the selection of 14 acres as the 2023 roundabout grapes vendor city staff would begin work with the vendor immediately to finalize a vendor agreement which would come before city council. Time is certainly of the essence the reason that we are bringing this forward tonight is because the grapes are apparently about ready to be picked there is a date chosen for picking i believe tomorrow morning. Yes. Volunteers will be out there picking grapes so. The purpose of this evening is to figure out what you want to do with the grapes you want to bring to my garage and we put them in there do we want to put them in city hall or do you want to actually work with somebody to make something out of them so that's what we're doing this evening and we do have dave regan from 14 acres is with us the winemaker and owner of 14 acres i spoke with him briefly he'd be happy to talk briefly about what his thoughts were on the partnership and what he might do with the 2023 grapes if you would like before you take action. . . . Hi folks dave regan 14 acres and. As you're correct you correctly heard that we are going to be picking tomorrow so you need to make a decision tonight okay you're going to decide you want them at your house or at the vineyard so yeah i'm the winemaker we bought three brothers winery i'm sure many people have heard of three brothers winery we bought that from the anderson's and. My wife and i rebranded it to 14 akers because it is actually on 14 acres that's the that's the surprise ending here. And and then about wine making i've been making wine for the last several years since i bought the vineyard i figured maybe i should probably make the wine here. We want a lot of gold awards with the la international the texas international and also the seattle international. Just in the last year so i think i've got more golden snoop dogg hanging on some of our bottles so kind of exciting stuff. . And to partner with the city and and. Also to just you know ron and i chat occasionally and i love helping the city out however i can regarding the grapes i mean. Even the manicuring of the vineyard. We do have eight acres of planted grapes. On our state but we also have. Another vineyard that we have a 10-year lease for in. Near the wazoo area and that's actually 6 acres. 14 acres of planted grapes. That we manage. The idea about picking tomorrow the weather's been awesome the grapes the sugar level and the grapes is excellent but. I think the one of the recommendations or one of the ideas is. Instead of making a read this year since we already have a red. Sitting for the city inside of barrel. Currently right now in my cellar. We could make a white. Roundabout rozay. You know it's kind of catchy. Add focus is a very aromatic grape so i anticipate would make a very aromatic wine. . With the with the fermentation and the production. I could have the wine ready bottled with labels on council's approval. By the end of september. . Yeah i can have that bottle by the at least the end of september labels would be the only thing that would hold it up which. . . And all that good stuff any questions so you're saying that this year's harvest of grapes would be. Ready to be bottled in september of next year this year. . . . I would i would gently suggest to put off the pic just so that we could. Get a little bit more sugar in their get a little bit more. Ripeness but seeing as how the city's already been doing red and i think there's a might be a plethora of red right now that you guys are sitting on. Might be a good idea to switch it up and do a rose and rosie's could be drink. Immediately first day it's bottled is the best day of its life. So let me kind of go down the line of the council and get your feedback on what you would like to do with this year's grapes if you want to do a do you want to do wine do you want to do something else and if we do wine do you want to do. Rose or. Counselor. . Here's the and by the way i checked it out today i did the bricks b r i x on the sugar content. All of the i did almost every row and they are all between 22 and 24. They they really ripen you know i went last monday to check them out. And call dave afterwards and said my god they're at 16 to 18. And they're ripening really fast this this is just happened in the last 2 weeks. 22 is the sweet spot. I think we ought to take a look at it and i think we ought to do a rose and dave. Just to give you a little bit of history dave has been out there helping. And he's talked to the other wineries to ask them for help on these roundabouts. And. Of course on the east side we had. It appears. Address of. . Keller herbicide. At the wiped out about half of the roundabout. The rest of them. The rest of the grapes on that east side were were. Looking bad. They've come back they're doing well and there are some grapes on that side now. But not a lot but the west side. . Melroy mr melroy and myself. Determine that we might go out there. With. . Scary the birds away if they show up and i'm afraid that they will show up this weekend. Like they did once before and then this this west side. Just loaded now. And part of it is because he came in. He didn't know that i was coming in with volunteers on a certain day i didn't know he was coming in with volunteers. The day before. And so he trimmed those grapes and they are ripening really fast so. First of all i think it's just great that we've got help. And i'd love to have work with dave again and i think it would be great to have a rosie why not. What are your thoughts. Having worked on the grapes for number of years i know if we don't. Quickly. We will lose all of them to the birds. That's happened two years in a row. I think we need to move and i would. . I agree on the rosie and we should pick tomorrow. Sounds good counselor wells. Rosie. . Rosie all day yesterday. I know that steve's. Garage would be a great place to put the trains absolutely. 14 acres okay but before we do that. We need to is there any further discussion about the grapes or possible vendor choice. My suggestion is that council adjourns and heads over to the winery and try some of the wine there before we you know you're a troublemaker okay so. Without any further discussion i would accept a motion. Adam are i moved that we accept the 14 acres to process the grapes. And this year in rosie style. Is there any further discussion. 14 acres winery as the vendor for the 2023 roundabout ridgeville roundabout grapes and to make a rose all those in favor say aye aye motion carries thank you and thank you so much thank you so much i look forward to working with you. . . So the last time a business this evening is resolution number 630 the closure of the large correctional center mr stuart thank you may remember to the community as you are aware a large correctional center is slated to be closed by the department of corrections in october of 2023 once notification of that closure was made many of the local jurisdictions in clark county have spoken up and out against that closure for a variety of reasons the city of richfield has specific reasons why we have benefited from and have benefited. The inmates in work crews from the large correctional facility that the council wanted to have their voice heard in this conversation to hopefully at least have that voice heard by the department of corrections and other state officials. Who made the ultimate determination about closure and. Seek some sort of. . . Other action to. Revoke and reject that direction and keep it open. So a resolution pro is prepared. For the council to sign that would be delivered to the state officials and two other relevant agencies the resolution reads whereas large correctional center is slated to be closed by the department of corrections in october 2023 and where is lcc. Provides valuable education retraining opportunities and community partnerships for inmates that reduce the recidivism rate of the inmates and provides valuable service to our community. And where is lcc inmate work crews play vital regional role in trail maintenance vegetation control and wildfire fighting and whereas lcc inmate work crews play a vital local role in ridgefield providing almost 11,000 hours since 2018. Assisting with roadside garbage pickup park trail roadway vegetation management and storm facility and rain garden maintenance and where is the residents of ridgefield have greatly benefited from work that couldn't be done by city public works crews alone. And save 4 3 4. 8 8. Due to the assistance of those lcc work crews and where is lcc inmates have benefited from the opportunity to work in and positively enhance communities in the region they're from. In ridgefield and is part of a retraining and education program which is greatly decreased recidivism. Now therefore be it resolved by the city council of the city of ridgefield washington as follows section 1 the city council recognizes the vital role that lcc work crews play regionally and locally aiding in park trail storm pond and right of way maintenance the city council strongly urges the department of corrections to reverse the decision to close large correctional facility. Section 2 the clerk of the city of richfield is hereby directed to forward copies of this resolution to the offices of the washington state department of corrections the governor and other such offices as applicable. Section 3 this resolution shall be effective immediately upon its adoption by city council. I'm done in open meeting of the city council of the city ridgefield washington held this 24th day of august 2023. Happy to answer any questions you might have. Thank you mr stuart are there any questions. Regarding the resolution. . What other cities have also done resolutions. The cities that end jurisdictions that i'm aware of in clark county is the county council for clark county issued a resolution the city council in battleground issued a resolution and the city council in the center. Issued a resolution of opposition to the closure those are the jurisdictions as i am aware of at this time. Any additional questions or comments cancel. Not seeing any i would accept emotion. 6:30 is presented. Is there a second. Is there any further discussion. Play we have a motion to adopt resolution number 630 as presented all those in favor say aye bye. Motion carries thank you council. We now have time again for public comments. Anyone wishing to address city council regarding items not subject to a specific public hearing may come forward at this time to the podium are used to raise him feature located on the toolbar at the bottom of your zoom screen. For controls over the phone press star 9 for raise hand and press star 6 to. Please state your name and limit your comments to 3 minutes is there anyone wishing to give. Public comment this evening. Is julia is there anyone online. . Not having anyone in person or online will move on to council reports. Cancel quickly let's do. Are reports and then we will move into executive sessions so counselor cole we'll start on your. We're in budget season full swing and i'm proud to be part of the budget advisory committee we had our first meeting last week and a presentation tonight it's always good and i think it's. A core function of this council is budget as well as policy so it's a real privileged participate in the process and to see how those things shape our city financially so i got to be a part of it thank you cancel it. So i just did a couple things just went to the tractor supply grand opening i was shocked at what a huge store it is i mean it's really a large store i think it's a great addition to richfield and then the wonderful farm to table dinner last saturday which was just a great event it's always a great event. And it was very fun so really enjoyed that. Thank you counselor wells. . I got three pages of things. They're little pages. . . . . . And. I didn't attend. Supply. Grand opening and the soft open. And i've got a comment. . Mr johnson. . We get the revenue. From tractor supply. You're going to have to commission. Hey loom. . Just take the money to the bank. There are people people were going in there right and left. It cost me $550. Scott jeffries my contractor. 800 bucks. And my son was there that night and he caught him 250 that was three nights ago at 8:00 at night so i mean there was just steady streams of people all the time. . . Thank you for supporting the local businesses. Was that a number of meetings this month but i just want to report on the one i went to last night was the ridgefield lions meeting. And during that community meeting they had reported the results of the income that was raised by the volunteers of 33,000 after all the bills are paid. And i know there was. 208 shifts filled by 53 lions there was multiple other entities that the high schools and the grade schools. The police department. The arts association the chamber of commerce shows all kinds of volunteers that help. I feel a great need and so that money will be used here in the community thank you. Thank you counselor burkle. Attended eight different functions. Clark regional wastewater budget review. When are the i want to report on. Program here in ridgefield called green bag food. . On 812. Wii. Is 8000 lb. First quality food. In ridgefield that remains in ridgefield going to neighbors helping neighbors. In the school resource. I think it's worthy of note. The program grow. . Time we do it. Only six times a year. . I was thrilled to death in addition. We collected 735 which will be used to buy. . Thank you. Great thank you so much. Cancel amazon i attended the grand opening of bigfoot my wife's birthday was the day before and so i spent a lot of money locally. Texarkana. . Antenna. Arcana. Stands for arkansas and montana. Unless i did not know that until the grand opening. Tractor supply was great farm to table with. Really great. Miss ireland. Is that her first or last name. . I can only remember some things so i remember ireland but anyway. It was really great i wasn't able to go last year and it was marvellous and we served the ridgefield round about red which i drink. And i've been working on the grapes every day this week because they are ripening so fast. And i think the birds will be here tomorrow so i mean saturday so i'm i'm glad that we're picking tomorrow. The one of note i am the chair of the c-tran in clark county and. They had a meeting with senator murray's staff two from washington dc and one from locally. And a tour of the new mill plain brt which by the way is the only bus rapid transit. In the united states that has. Installed and done. . Within the budget under budget. And on time so they were very impressed with that and senator murray really does help quite a bit. In our area. . Thank you i just want to give you things out to everybody on the council i was gone a lot of last week taking my daughter to college and specialty out. With all these ribbon cutting all these new businesses they all i think. But i did have time for. One really important meeting and it kind of piggybacks on to what cancel was saying with senator murray staffing in town. I was able to take around on a tour of richfield. He had not been out to ridgefield in many years and was. Surprise surprise impressive there's a lot of changes here and i will have the opportunity to take him around and show him like where the 11th street overpass is going to be. . . The well up and you know at kennedy farms in several other projects and the pioneer street east extension so it was really great that we had our local staffer and our. The federal. . . Get some traction on those projects at the national level. With that and we are going to be moving into executive session. For the city manager performance evaluation pursuant to rcw 42.30 dot 110 subsection 1 z. . We will be in executive session until 8:00. An action will be taken when we return so cancel. . . Adjourn to executiv"

# Extract a longer summary (e.g., 5 sentences)
summary = extract_key_points(text, num_sentences=5)
print(summary)
```
